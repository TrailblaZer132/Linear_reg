```python
#making the imports
import pandas as pd #making calculations
import numpy as np #read data
import matplotlib.pyplot as plt #plotting graph
plt.rcParams['figure.figsize'] = (10.0, 9.0) #determining the plot size

#read data
data = pd.read_csv('test.csv')
X= data.iloc[:,0]
y= data.iloc[:,1]
plt.scatter(X,y)
plt.show()
```

```python
#Building the model
m= 0
c= 0

L= 0.0001
epochs= 1000

n= float(len(X))

#Gradient descent
for i in range(len(X)):
    y_pred= m * X + c
    D_m= (-2/n) * sum(X*(y - y_pred))
    D_c= (-2/n) * sum(y - y_pred)
    
    #updating the values of m and c
    m = m - (L * D_m)
    c = c - (L * D_c)
    
print(m, c)
```

```python
#making predictions
y_pred= m * X + c

plt.scatter(X,y)
plt.scatter(X,y_pred)
plt.show()
```

```python
plt.scatter(X,y,color='pink')
plt.plot([min(X),max(X)], [min(y_pred), max(y_pred)], color='blue')
plt.show()
```
